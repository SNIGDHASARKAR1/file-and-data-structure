17.  Describe the time complexity of accessing elements in a dictionary.

     -   Accessing elements in a Python dictionary is highly efficient due to the underlying implementation of hash tables. Let's break down the time complexity:

     Average Case:
     Time Complexity: O(1)

     Reason: In the average case, accessing an element by its key involves computing the hash value of the key, which directs you to the exact location in the hash table where the value is stored. This direct access results in constant time complexity.

     Worst Case:
     Time Complexity: O(n)

     Reason: In the worst case, all keys hash to the same index, leading to collisions. Python handles collisions using a technique called "open addressing," where it searches for the next available slot. If all keys end up in the same slot, it might need to search through all elements, resulting in linear time complexity.
